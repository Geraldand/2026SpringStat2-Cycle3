### Checking Assumptions for Two-Proportion Z-test
Before running the test, we must check the conditions:
1. **Independence:** The two groups (felt sad/hopeless vs. did not) are independent.
2. **Success/Failure Condition:** Both groups must have at least 10 successes and 10 failures.

```python
import pandas as pd

# 因為檔案直接傳在左邊，這裡路徑直接寫檔名即可
df = pd.read_csv('cleaned_data.csv')

# 建立交叉表來檢查數量
assumption_check = pd.crosstab(
    df['SadOrHopeless'],
    df['CurrentCigaretteUse'],
    margins=True,
    margins_name="Total"
)

print("Count of Successes and Failures:")
display(assumption_check)
```
