## PART 1: KEY STATISTICAL FINDINGS (基礎核心統計推論)
**Dataset:** YRBS 2007 (Valid Sample Size: $N = 13,174$)

* **Group 1 (Sad or Hopeless):** 27.61% Smoking Rate
* **Group 2 (Not Sad):** 16.18% Smoking Rate
* **Net Absolute Difference:** +11.43%
* **Statistical Inference:** $Z = 15.0536$, $P < 0.001$ (Strongly Reject $H_0$)
* **95% Confidence Interval:** $[9.83\%, 13.02\%]$

**[Conclusion]** Students who felt sad or hopeless have a significantly higher smoking rate. Visual evidence confirms that the theoretical sampling distributions are completely separated, proving this 11.43% gap is robust and not due to chance.  

*(感到悲傷或絕望的學生，吸菸率顯著高於無此情緒者。兩群體的差距高達 11.43%，統計結果極度顯著。)*

## PART 2: THE DEEP DIVE - EDA & COMPOUNDING RISKS (進階探索與疊加陷阱)
Exploratory Data Analysis (EDA) reveals that smoking is deeply intertwined with mental health and substance abuse, forming a severe "Risk Escalator."

*(透過進階資料探索，我們發現吸菸行為並非單一現象，而是呈現顯著的「共病風險階梯」：)*

1. **The Compounding Trap (7.8x Higher Risk / 7.8倍的複合風險):** * **Baseline:** Students with NO sadness and NO binge drinking have a smoking rate of only **7.0%**.
   * **Comorbidity:** When sadness is compounded with binge drinking, the smoking rate violently skyrockets to **54.8%** (a **7.8-fold** increase from the baseline).

   *(當正常無風險的基準吸菸率僅有 7.0% 時，若同時疊加「悲傷」與「狂飲」兩大負面因子，吸菸率將暴漲至 54.8%，風險高達 7.8 倍。)*

2. **The Beacon of Sadness (42.8% 悲傷濃度信標):** * Reversing the perspective, students who exhibit both smoking and binge drinking behaviors demonstrate a severe depression rate of **42.8%**, far exceeding the school-wide average (~29%).

   *(視角反轉後發現，同時具備吸菸與狂飲行為的學生群體中，感到嚴重悲傷的比例飆升至 42.8%，遠超全校基準。)*

3. **Shattering Rebellion Stereotypes (破除純粹叛逆迷思):** * Only **20.1%** of student smokers fit the "pure rebellion" stereotype.
   * The vast majority (**79.9%**) carry invisible burdens of psychological distress or alcohol dependency. Smoking serves as a critical "radar beacon" for trauma, not merely disciplinary misconduct.

   *(僅 20.1% 為純粹叛逆型吸菸。高達近八成的吸菸青少年皆伴隨心理困擾或酒精濫用。吸菸行為已非單純紀律問題，而是內心創傷的求救信標。)*

## 🏛️ PART 3: ACTIONABLE POLICY IMPLICATION (政策呼籲)
School campaigns must pivot from "behavior isolation and punishment" to "Integrated Care Models." To truly reduce youth smoking, educational institutions must combine Tobacco Cessation with active Student Mental Health Support.

*(校園菸害防制必須從「單純行為懲罰」轉型。必須將「戒菸輔導」與「心理健康支持」無縫結合，才能真正接住這些隱藏在菸霧背後的無聲求救。)*
